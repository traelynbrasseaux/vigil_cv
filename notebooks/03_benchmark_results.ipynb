# 03 - Benchmark Results

Load and visualize benchmark results across backends and precision modes.

```python
import sys
sys.path.insert(0, "..")

from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the most recent benchmark CSV
results_dir = Path("../benchmark/results")
csv_files = sorted(results_dir.glob("benchmark_*.csv"))

if not csv_files:
    print("No benchmark results found. Run benchmark/run_benchmark.py first.")
else:
    df = pd.read_csv(csv_files[-1])
    print(f"Loaded: {csv_files[-1].name}")
    display(df)
```

```python
if csv_files:
    fig, axes = plt.subplots(1, 3, figsize=(16, 5))

    labels = [f"{r['backend']}\n{r['precision']}" for _, r in df.iterrows()]

    # Latency
    axes[0].bar(labels, df["mean_latency_ms"], color="#2196F3")
    axes[0].set_ylabel("Latency (ms)")
    axes[0].set_title("Mean Latency")

    # FPS
    axes[1].bar(labels, df["fps"], color="#FF9800")
    axes[1].set_ylabel("FPS")
    axes[1].set_title("Throughput")

    # AUROC
    axes[2].bar(labels, df["auroc"], color="#4CAF50")
    axes[2].set_ylabel("AUROC")
    axes[2].set_title("Detection Accuracy")
    axes[2].set_ylim(0.8, 1.0)

    plt.tight_layout()
    plt.show()
```

```python
if csv_files:
    # Latency distribution
    fig, ax = plt.subplots(figsize=(10, 5))

    x = range(len(df))
    width = 0.2

    ax.bar([i - width for i in x], df["p50_latency_ms"], width, label="P50", color="#4CAF50")
    ax.bar(x, df["p95_latency_ms"], width, label="P95", color="#FF9800")
    ax.bar([i + width for i in x], df["p99_latency_ms"], width, label="P99", color="#F44336")

    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.set_ylabel("Latency (ms)")
    ax.set_title("Latency Percentiles by Backend")
    ax.legend()

    plt.tight_layout()
    plt.show()
```
