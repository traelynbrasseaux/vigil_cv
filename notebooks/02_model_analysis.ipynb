# 02 - Model Analysis

Analyze backbone feature extraction, PatchCore memory bank characteristics,
and compare detection quality across backbones.

```python
import sys
sys.path.insert(0, "..")

from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import torch
from models.backbone import FeatureExtractor
```

```python
# Compare backbone architectures
for name in ["efficientnet", "mobilenet"]:
    model = FeatureExtractor(backbone_name=name)
    x = torch.randn(1, 3, 224, 224)
    with torch.no_grad():
        out = model(x)
    params = sum(p.numel() for p in model.parameters())
    print(f"{name}: output={out.shape}, params={params:,}")
```

```python
# Visualize feature maps
model = FeatureExtractor(backbone_name="efficientnet")
x = torch.randn(1, 3, 224, 224)
with torch.no_grad():
    features = model(x)

fig, axes = plt.subplots(2, 4, figsize=(12, 6))
fig.suptitle("EfficientNet-B0 Feature Maps (first 8 channels)", fontsize=12)

for i in range(8):
    ax = axes[i // 4, i % 4]
    ax.imshow(features[0, i].numpy(), cmap="viridis")
    ax.set_title(f"Channel {i}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Autoencoder reconstruction analysis
from models.autoencoder import ConvAutoencoder

ae = ConvAutoencoder()
x = torch.randn(1, 3, 224, 224)
with torch.no_grad():
    recon = ae(x)
    score, heatmap = ae.anomaly_score(x)

print(f"Reconstruction shape: {recon.shape}")
print(f"Anomaly score: {score.item():.4f}")
print(f"Heatmap range: [{heatmap.min():.4f}, {heatmap.max():.4f}]")
```
