# 01 - Data Exploration

Explore the MVTec Anomaly Detection dataset structure, visualize samples,
and understand the distribution of normal vs. anomalous images.

```python
import sys
sys.path.insert(0, "..")

from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import torch
from data.dataset import MVTecDataset, IMAGENET_MEAN, IMAGENET_STD
```

```python
DATA_ROOT = Path("../data/mvtec_anomaly_detection")
CATEGORY = "leather"

train_dataset = MVTecDataset(root=DATA_ROOT, category=CATEGORY, split="train")
test_dataset = MVTecDataset(root=DATA_ROOT, category=CATEGORY, split="test")

print(f"Train samples: {len(train_dataset)}")
print(f"Test samples: {len(test_dataset)}")
```

```python
def denormalize(tensor):
    """Reverse ImageNet normalization for visualization."""
    mean = torch.tensor(IMAGENET_MEAN).view(3, 1, 1)
    std = torch.tensor(IMAGENET_STD).view(3, 1, 1)
    return (tensor * std + mean).clamp(0, 1)

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle(f"MVTec AD - {CATEGORY}", fontsize=14)

# Show train samples (normal)
for i in range(5):
    img, label, mask, cat = train_dataset[i]
    axes[0, i].imshow(denormalize(img).permute(1, 2, 0).numpy())
    axes[0, i].set_title(f"Train (normal)")
    axes[0, i].axis("off")

# Show test samples (mix of normal and anomalous)
for i in range(5):
    img, label, mask, cat = test_dataset[i]
    status = "anomaly" if label == 1 else "normal"
    axes[1, i].imshow(denormalize(img).permute(1, 2, 0).numpy())
    axes[1, i].set_title(f"Test ({status})")
    axes[1, i].axis("off")

plt.tight_layout()
plt.show()
```

```python
# Label distribution
test_labels = [test_dataset[i][1] for i in range(len(test_dataset))]
normal_count = test_labels.count(0)
anomaly_count = test_labels.count(1)

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(["Normal", "Anomalous"], [normal_count, anomaly_count], color=["#4CAF50", "#F44336"])
ax.set_ylabel("Count")
ax.set_title(f"Test Set Distribution - {CATEGORY}")
plt.tight_layout()
plt.show()
```
